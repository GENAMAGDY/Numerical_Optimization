## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate. 

```python
Theta = [0,0]
Learning_rate = 0.001
m=len(X)
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
Hypotheses = Theta[0] + Theta[1]*X
Hypotheses
```

### Step3: Calculate Cost function ùë±(theta_0,theta_1 ).


```python
CostFunction= (1/(2*m)) * np.sum((Hypotheses - y)**2)
CostFunction
```

### Step4: Calculate the gradient.


```python
grad =[0,0]
grad[0] = (1/m)*np.sum(Hypotheses - y)
grad[1]= (1/m)*np.sum(np.dot((Hypotheses - y),X))
grad
```

### Step5: Update the parameters (simultaneously).


```python
Theta[0] -= Learning_rate*grad[0]
Theta[1] -= Learning_rate*grad[1]
Theta
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.

```python
Theta = [0,0]
Learning_rate = 0.001
for i in range(1000):
    Hypotheses = Theta[0] + Theta[1]*X
    CostFunction= (1/(2*m)) * np.sum((Hypotheses - y)**2)
    grad =[0,0]
    grad[0] = (1/m)*np.sum(Hypotheses - y)
    grad[1]= (1/m)*np.sum(np.dot((Hypotheses - y),X))
    Theta[0] -= Learning_rate*grad[0]
    Theta[1] -= Learning_rate*grad[1]
    print(Theta)
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x

```python
y_pre = Theta[0]+Theta[1]*X
y_pre
```

#### Plot  LR equation output (fitted line) with the original data (scatter plot of X,y)

```python
plt.scatter(X,y)
plt.plot(X,y_pre ,color='Red')
plt.show()
```

#### Use R2 score to evaluate LR equation output



https://en.wikipedia.org/wiki/Coefficient_of_determination

```python
from sklearn.metrics import r2_score
score=r2_score(y_pre , y)
score
```
